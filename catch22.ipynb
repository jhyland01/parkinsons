```python
import pycatch22
import os
import pandas as pd
cpd = pd.read_csv('./cpd_data/filtered_data.csv')
unique_series_ids = cpd['SeriesId'].unique()

# Whether to compute the mean and standard deviation:
doCatch24 = True

```

```python
import numpy as np
# Extract features from each data file in testData
for seriesId in unique_series_ids:

    print ('\n'), seriesId

    subset = cpd[cpd['SeriesId'] == seriesId]

    # drop the SeriesId column
    subset = subset.drop(['SeriesId'], axis=1)

    # turn into numpy array
    subset = subset.to_numpy()

    flat_data = [float(item) for sublist in subset for item in sublist]

    catchOut = pycatch22.catch22_all(flat_data, catch24 = doCatch24)

    featureNames = catchOut['names']
    featureValues = catchOut['values']

    for featureName, featureValue in zip(featureNames, featureValues):
        print('%s : %1.6f' % (featureName, featureValue))
```
